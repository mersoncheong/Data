# Storytelling Data Visualization on Exchange Rates

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

Daria Chemkaeva put together the data set and made it available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) — the data source is the European Central Bank. Note that the dataset gets regular updates — we downloaded it on January 2021.

Let's begin by reading in the dataset. 

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

Before, we proceed on with the data visualisation, we will conduct some data cleaning.

## Data Cleaning

The focus of this project will be on the exchange between the euro and American Dollar. Hence we will isolate the data needed for us to achieve our aim.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
# Remove rows where - character appears
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

The exchange rate is subjected to constant movement on a daily basis. As our goal is to see the long term trend, we shall use a rolling mean of 30 days to construct a smoother graph for us to more clearly identify any patterns.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar.rolling(30).mean()
```

## Idea

Now, we will use the data obtained to show how the euro-dollar rate has changed during the recent coronavirus pandemic. We will use the 2020 data as 2016-2019 as baseline and visualize it through a lineplot.

## Graph Sketching

```python
#Isolate 2016 to 2021 time data
euro_to_dollar = euro_to_dollar[euro_to_dollar['Time'].dt.year >= 2016]
```

```python
euro_to_dollar.info()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline
style.use('fivethirtyeight')
```

```python
fig,ax = plt.subplots(figsize=(9,3))
# Plot Graph
ax.plot(euro_to_dollar['Time']
        ,euro_to_dollar['rolling_mean']
       ,color='Red'
        ,linewidth=2
        ,alpha = 0.1)
# Highlight time of Pandemic
ax.plot(euro_to_dollar['Time'][1087:1284]
        ,euro_to_dollar['rolling_mean'][1087:1284]
       ,color='Red'
       ,linewidth=3)

# Add Title
ax.text(735999,1,'How the Coronavirus Pandemic has affected the Euro-USD Exchange Rate',
       weight = 'bold',size=12
    )
# Add subtitle
ax.text(736375,0.98,'Euro-USD exchange rates between 2016 and 2021',size = 11)
ax.text(737450,1.074,'Outbreak of Coronavirus in US',size =6)

# Add Signature
ax.text(735700,0.95,
        '©Merson Graph' + ' '*70+ 'Source: European Central Bank',
         color = '#f0f0f0',
        backgroundcolor = '#4d4d4d')

plt.show()
```
